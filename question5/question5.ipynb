```python
import numpy as np
import matplotlib.pyplot as plt
```

$$\frac{d}{dx}x(x-1)=(x-1)+x=2x-1$$
Hence, $$\frac{d}{dx}x(x-1)\Big|_{x=1}=1$$

```python
# Compute the derivative of function f at point x numerically.
def d(f, x, delta):
    return (f(x+delta)-f(x))/delta
```

```python
def f(x):
    return x*(x-1)
```

```python
delta = np.logspace(-4,-14,1000)
plt.semilogx(delta, d(f, 1, delta))
plt.xlabel(r'$\delta$', fontsize = 15)
plt.ylabel(r'$\frac{f(1+\delta)-f(1)}{\delta}$', fontsize = 20)
plt.title(r'Numerical derivative of $f(x)=x(x-1)$ at $x=1$', fontsize = 14)
plt.grid()
```

It appears that the numerical approximation of the derivative starts to rapidly accumulate error as $\delta$ becomes very small. I presume that floating point errors become more significant for very small values of $\delta$, thus causing the error accumulation.
